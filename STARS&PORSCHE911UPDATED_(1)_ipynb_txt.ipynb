```python

from typing import Optional
import torch
from llama_index.core import Settings, VectorStoreIndex, Document
from llama_index.embeddings.langchain import LangchainEmbedding
from langchain.embeddings.huggingface import HuggingFaceEmbeddings
from transformers import AutoTokenizer, AutoModelForCausalLM
from dataclasses import dataclass

@dataclass
class LLMConfig:
    context_window: int = 4096
    max_new_tokens: int = 4096
    chunk_size: int = 512
    model_name: str = "all-MiniLM-L6-v2"
    device: str = "cuda" if torch.cuda.is_available() else "cpu"

class StarbaseAssistant:
    def __init__(
        self,
        model,
        tokenizer,
        config: Optional[LLMConfig] = None
    ):
        self.config = config or LLMConfig()
        self.model = model
        self.tokenizer = tokenizer
        self.setup_system()

    def setup_system(self):
        """Initialize the system with proper prompts and settings"""
        self.system_prompt = """<s>[INST] <<SYS>>
        You are a helpful, respectful, and honest assistant specializing in astronomical knowledge.
        Always answer as helpfully as possible while maintaining accuracy and safety.
        Your answers should be focused on astronomical phenomena and celestial objects based on the provided dataset.
        If a question is unclear or lacks factual coherence, explain why rather than providing incorrect information.
        If you're unsure about an answer, acknowledge the limitation rather than making assumptions.
        <</SYS>>"""

        self.query_wrapper_prompt = "{query_str}"

        # Setup embeddings
        self.embeddings = LangchainEmbedding(
            HuggingFaceEmbeddings(model_name=self.config.model_name)
        )

        # Configure global settings
        Settings.llm = self.create_llm()
        Settings.embed_model = self.embeddings
        Settings.chunk_size = self.config.chunk_size

    def create_llm(self):
        """Create a HuggingFaceLLM instance with proper configuration"""
        from llama_index.llms import HuggingFaceLLM

        return HuggingFaceLLM(
            context_window=self.config.context_window,
            max_new_tokens=self.config.max_new_tokens,
            system_prompt=self.system_prompt,
            query_wrapper_prompt=self.query_wrapper_prompt,
            model=self.model,
            tokenizer=self.tokenizer,
            device=self.config.device
        )

    def create_index(self, documents: list[Document]) -> VectorStoreIndex:
        """Create a vector store index from documents"""
        return VectorStoreIndex.from_documents(documents)

    def generate_response(self, query_text: str) -> str:
        """Generate a response for the given query"""
        input_tokens = self.tokenizer(
            query_text,
            return_tensors="pt"
        ).to(self.config.device)

        output = self.model.generate(
            **input_tokens,
            max_new_tokens=self.config.max_new_tokens,
            do_sample=True,
            temperature=0.7,
            top_p=0.9
        )

        return self.tokenizer.decode(output[0], skip_special_tokens=True)

# Usage example:
def setup_assistant(model_path: str):
    """Setup the StarbaseAssistant with the specified model"""
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model = AutoModelForCausalLM.from_pretrained(
        model_path,
        device_map="auto",
        torch_dtype=torch.float16
    )

    return StarbaseAssistant(model=model, tokenizer=tokenizer)

# Example usage:
# assistant = setup_assistant("your_model_path")
# index = assistant.create_index(your_documents)
# query_engine = index.as_query_engine()
# response = assistant.generate_response("Tell me about the stars")
```
